### Calidad de datos

```python
def revisar_calidad_de_datos(data, dataNombre):
    '''
    Revisa la calidad de los datos devolviendo un reporte de datos nulos, tiposd e datos y valores unicos por columna
    '''
    print(f"------------- Dataset de {dataNombre} -------------")
    duplicados = data.duplicated().sum()
    print(f"Cantidad de duplicados: {duplicados}")
    print("----------------------------------------")
    tipos_de_datos = data.dtypes;
    nulos = data.isnull().sum();
    porcentaje_nulos = (nulos / len(data)) *100;
    unicos = data.nunique();

    reporte = pd.DataFrame({
        "Tipo de dato": tipos_de_datos,
        "Cantidad de nulos": nulos,
        "Porcentaje de nulos": porcentaje_nulos,
        "Cantidad de valores unicos": unicos
    })
    return reporte;

```

```python
revisar_calidad_de_datos(clientes, "Clientes")
```

```python
revisar_calidad_de_datos(ventas, "Ventas")
```

```python
revisar_calidad_de_datos(marketing, "Marketing")
```

### Limpieza de texto

```python
def normalizar_texto(texto):
  '''
  Convertir texto a minusculas, quitar tildes y espcacios extras
  '''
  if not isinstance(texto, str):
    return texto;
  texto = texto.lower();
  reemplazos = {
        "√°": "a", "√©": "e", "√≠": "i", "√≥": "o", "√∫": "u",
        "√§": "a", "√´": "e", "√Ø": "i", "√∂": "o", "√º": "u",
        "√±": "n"
  }
  for acento, sin_acento in reemplazos.items():
      texto = texto.replace(acento, sin_acento);
  texto = texto.strip();
  return texto;


```

### Limpieza general de datos

```python
def limpiar_datos(data, columnas=None):
  data = data.copy();
  data = data.drop_duplicates();
  if columnas:
    for columna in columnas:
      if columna in data.columns: data[columna] = data[columna].astype(str).apply(normalizar_texto)
  return data;

cliente_limpio= limpiar_datos(clientes, columnas=['nombre', 'ciudad'])
venta_limpio = limpiar_datos(ventas, columnas=['producto', 'categoria'])
marketing_limpio = limpiar_datos(marketing, columnas=['canal_venta', 'canal'])

print("-----------------------------------------------------------------------")
print("limpieza completa.")
print("------------------------------- Clientes -------------------------------")
display(cliente_limpio.head())
print("------------------------------- Ventas -------------------------------")
display(venta_limpio.head())
print("------------------------------ Marketing -------------------------------")
display(marketing_limpio.head())
```

### Transformacion de datos

```python
if 'monto' in venta_limpio.columns:
  p80 = venta_limpio['monto'].quantile(0.8)
  top_ventas = venta_limpio[venta_limpio['monto'] >= p80]
  print(f"Producto con ventas mayot igual a percentil 80 ({p80:.2f}) ")
  print(top_ventas.head())

```

### Agregacion

```python
if {'producto', 'monto'}.issubset(venta_limpio.columns):
    resumen_ventas = (
        venta_limpio.groupby('producto')
                     .agg(total_ventas=('monto', 'sum'),
                          promedio=('monto', 'mean'),
                          cantidad=('monto', 'count'))
                     .reset_index()
                     .sort_values('total_ventas', ascending=False)
    )
    display(resumen_ventas.head())
```
